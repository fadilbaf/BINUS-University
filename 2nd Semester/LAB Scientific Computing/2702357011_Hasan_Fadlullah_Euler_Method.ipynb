```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
# Define the ODE
f = lambda x, y: x**2 - 4*y

# Step size
h = 0.01

# Numerical grid
t = np.arange(0, 1 + h, h)

# Initial condition
y0 = 1

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

# Exact solution for the ODE y' + 4y = x^2
exact_solution = lambda x: (x**2/4 - x/8 + 1/64) + (63/64) * np.exp(-4 * x)

# Compute the exact solution values
exact_y = exact_solution(t)

# Plotting the results
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Approximate (Euler Method)')
plt.plot(t, exact_y, 'g', label='Exact')
plt.title('Approximate and Exact Solution for the ODE y\' + 4y = x^2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

# Find the value of the approximate solution at t = 0.1
t_index = np.where(t == 0.1)[0][0]
approx_value_at_0_1 = y[t_index]

print(f'The value of the approximate solution at t = 0.1 is {approx_value_at_0_1:.4f}')
```
